```python
import requests
from bs4 import BeautifulSoup

r = requests.get("http://www.oabrj.org.br/")
c = r.content

c
```

```python
soup=BeautifulSoup(c,"html.parser")
```

```python
all=soup.find_all("div",{"class":""})
```

```python
all
```

```python
import pandas as pd

df= pd.DataFrame(all)
df.to_dict()

df
```

```python
df.to_csv('all.csv', index=False)
print('Done')
```
