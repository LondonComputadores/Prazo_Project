```python
import requests
from bs4 import BeautifulSoup

r = requests.get("http://www.diariooficial.com.br/licitacoes.php")
c = r.content

c
```

```python
soup=BeautifulSoup(c,"html.parser")
```

```python
arquivo_licitac2 = soup.select("body a")
```

```python
arquivo_licitac2
```

```python
import pandas
df = pandas.DataFrame(arquivo_licitac2)
```

```python
df.to_csv("arquivo_licitac2.csv", index=False, encoding='utf-8', header=None)
```
